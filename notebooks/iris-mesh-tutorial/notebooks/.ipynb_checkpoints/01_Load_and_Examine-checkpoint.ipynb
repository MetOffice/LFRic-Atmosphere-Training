# Section 1: Loading and Examining LFRic and UM data

This section uses output from an LFRic run and a UM-style lat-lon run to show practical differences between unstructured mesh data and structured grid data in Iris.


```python
import warnings
import iris

# Some datasets trigger repeated NumPy masked-array cast warnings during load.
# These are expected for this tutorial data and can obscure the key learning output.
warnings.filterwarnings(
    'ignore',
    message='invalid value encountered in cast',
    category=RuntimeWarning,
    module=r'numpy\.ma\.core',
)

# Iris warns that date precision defaults are changing; opt in now for stable behaviour.
iris.FUTURE.date_microseconds = True
```

## Load LFRic and UM data from netCDF files
The tutorial helper code [testdata_fetching](./support/testdata_fetching.py) defines file paths for both datasets:

- `lfric_filepth`: unstructured UGRID/netCDF output
- `um_filepth`: structured lat-lon style netCDF output


```python
# import tutorial helper routines for handling access to test data
from support.testdata_fetching import lfric_filepth, um_filepth
print('LFRic data availble from ' + str(lfric_filepth))
print('UM data availble from ' + str(um_filepth))
```

We use [iris.load](https://scitools-iris.readthedocs.io/en/latest/userguide/loading_iris_cubes.html) for both files.

For the LFRic UGRID file, we use `iris.LOAD_POLICY.context(support_multiple_references=True)` so Iris correctly handles shared mesh metadata references during load.


```python
print('Loading LFRic and UM datasets...')

with iris.LOAD_POLICY.context(support_multiple_references=True):
    lfric_cubes = iris.load(lfric_filepth)

um_cubes = iris.load(um_filepth)

print(f'Loaded {len(lfric_cubes)} LFRic cubes from: {lfric_filepth}')
print(f'Loaded {len(um_cubes)} UM cubes from: {um_filepth}')

print('\nExample LFRic cube:')
print(lfric_cubes[0])
print('\nExample UM cube:')
print(um_cubes[0])
```

## What to compare between LFRic and UM cubes
When you inspect cubes, focus on these differences:

- **LFRic mesh cube**: includes mesh-related metadata (`mesh`, `location`, and `mesh_dim`) because data is attached to mesh elements (for example faces).
- **UM-style structured cube**: does not use mesh metadata; it is described by regular coordinate dimensions (for example latitude/longitude).

Why this matters:

- Mesh metadata is essential for operations like mesh-aware plotting, region extraction, and mesh regridding.
- Structured-grid workflows usually operate directly on coordinate dimensions and their bounds.

In other words, both are `Cube` objects in Iris, but their spatial description model is different.

## Loading a single cube
To inspect mesh-specific properties explicitly, load one LFRic variable and print:

- `cube.mesh`
- `cube.location`
- `cube.mesh_dim()`


```python
with iris.LOAD_POLICY.context(support_multiple_references=True):
    lfric_rh = iris.load_cube(lfric_filepth,'relative_humidity_wrt_water')

print(lfric_rh)
print(lfric_rh.mesh)
print(lfric_rh.location)
print(lfric_rh.mesh_dim)
```

If a cube is not mesh-based, these properties are `None` (or not meaningful for mesh operations).
The following UM cube example illustrates that contrast:


```python
um_cube = iris.load_cube(um_filepth,'air_temperature')
print(um_cube.mesh)
```
